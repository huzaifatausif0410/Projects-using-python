```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df=pd.read_excel('cereals_practice.xlsx')
df
```

```python
a=pd.get_dummies(df[['mfr', 'type']])
a
```

```python
df1 = pd.concat([df, a], axis=1, sort=False)
df1
```

```python
df1.drop(columns=['name','mfr','type'], axis=1,inplace=True)
df1
```

```python
df1.fillna(df1.mean(), inplace=True)
df1
```

```python
df1.dtypes
```

```python
df1.astype('float64').dtypes
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
X = df1.values
X = np.nan_to_num(X)
clus_dataset = StandardScaler().fit_transform(X)
clus_dataset
```

```python
from sklearn.cluster import KMeans 
clusterNum = 5
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 55)
k_means.fit(X)
k_means.labels_
```

```python
df["Clus_label"] = k_means.labels_
df
```

```python
k_means.inertia_
```

```python
df=df[['name',
 'Clus_label',
 'mfr',
 'type',
 'calories',
 'protein',
 'fat',
 'sodium',
 'fiber',
 'carbo',
 'sugars',
 'potass',
 'vitamins',
 'shelf',
 'weight',
 'cups',
 'rating']]
df
```

```python
df.to_excel('cereals_data_clustered.xlsx')
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df['Clus_label'], X[:,12], c=k_means.labels_.astype(np.float))
plt.xlabel('Cluster-Label', fontsize=15)
plt.ylabel('Rating', fontsize=16)
plt.show()
```

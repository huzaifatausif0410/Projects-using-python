```python
link='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
import pandas as pd
```

```python
df1=pd.read_html(link)
df=df1[0]
df=df[df['Borough'] != 'Not assigned'].reset_index() #discarding the values which are 'Not Assigned'
df
```

```python
df=df.groupby("Postcode").agg(lambda x:','.join(set(x))) #Grouping values that have same 'Postcode'
df.reset_index()
```

```python
df['Neighbourhood'].replace("Not assigned", df["Borough"], inplace=True) # Replacing 'Not Assigned' 'Neighbourhood' values with 'Bourugh' values.
df=df.reset_index()
df
```

```python
df.shape
```

```python
coords=pd.read_csv('http://cocl.us/Geospatial_data') # reading geospatial data into dataframe
coords
```

```python
df=df.merge(coords, left_on='Postcode', right_on='Postal Code') # merging the two dataframes
df=df.drop(columns='Postal Code', axis=1) # deleting unnecesary column/s
df
```
